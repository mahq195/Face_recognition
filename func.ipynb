```python
import os
import glob
from distutils.dir_util import copy_tree
```

```python
def move(original, target):
    import shutil
    shutil.move(original, target)
```

```python
path = r'lfw1000'
parent_path = r'raw_data'
sub_dir = os.listdir(path)


for sdir in sub_dir:
    sub_dir_path = os.path.join(path, sdir)
    image_path = os.listdir(sub_dir_path) 
    print(image_path)

    dst_path = os.path.join(os.path.join(parent_path, sdir) , image_path[0])
    print(dst_path)
    os.mkdir(os.path.join(parent_path, os.path.basename(sub_dir_path)))
    src = os.path.join(os.path.join(path, sdir), image_path[0])
    move(src, dst_path)
    

    
```

```python
full_lfw = r'lfw-deepfunneled'
lfw1000 = r'lfw1000'

dir_list = glob.glob(full_lfw + '\\*')

count = 0
for adir in dir_list:
    # print(adir)
    name = os.path.basename(adir)
    dst = os.path.join(lfw1000, name)
    
    img_ls = os.listdir(adir)
    print(len(img_ls))
    if len(img_ls) > 1:
        if not os.path.exists(dst):
            os.mkdir(dst)
            print('mk')
        copy_tree(adir, dst)
        count += 1
        
    if count == 1000:
        break
        
```
